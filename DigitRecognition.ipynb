```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("A_Z Handwritten Data.csv")
```

```python
df.shape
```

```python
df.tail(20)
```

```python
X = df.drop('0',axis=1)
y = df['0']
X.shape
```

```python
img = X.iloc[126010,:].values.reshape(28,28)
plt.imshow(img,cmap='gray')
```

```python
X_scaled = X/255
```

```python
from tensorflow.keras.utils import to_categorical

ya = to_categorical(y,num_classes=26)
```

```python
ya.shape
```

# Model Building

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
model = Sequential()
model.add(Dense(16,activation='relu',input_shape=(784,)))
model.add(Dense(16,activation='relu'))
model.add(Dense(26,activation='softmax'))
model.compile(loss='categorical_crossentropy',metrics=['accuracy'])
X_subset = X_scaled[:100000]
y_subset = ya[:100000]
model.fit(X_subset, y_subset, epochs=30, batch_size=64)
```

```python
yp = model.predict(X_scaled.iloc[:4,:])
yp.argmax(axis=1)
```

```python
def get_digit(filename):
    img = cv2.imread('../../DSML36/Deep Learning/A_image/'+filename,0)
    
    img = cv2.resize(img,(28,28))
    # plt.imshow(img,cmap='gray')
    img = img/255
    return model.predict_on_batch(img.reshape(1,784)).argmax()
    # X_scaled.shape
    
```
